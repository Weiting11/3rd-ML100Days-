```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = './data/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
#for test
unique_house_type = app_train['HOUSETYPE_MODE'].unique()
unique_house_type
amt_data = app_train[['TARGET', 'AMT_CREDIT']]
amt_data.head()
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
unique_house_type = app_train['HOUSETYPE_MODE'].unique()
amt_data = app_train[['TARGET', 'AMT_CREDIT']]

nrows = len(unique_house_type)
ncols = nrows // 2 #i.e 50/2=25.0 ; 50//2=25 (integer)

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    
    sns.distplot(app_train.loc[(app_train['HOUSETYPE_MODE'] == unique_house_type[i]) & \
                              (amt_data['TARGET'] == 0), 'AMT_CREDIT'], label = str(unique_house_type[i]),kde=False)
    sns.distplot(app_train.loc[(app_train['HOUSETYPE_MODE'] == unique_house_type[i]) & \
                              (amt_data['TARGET'] == 1), 'AMT_CREDIT'], label = str(unique_house_type[i]),kde=False)

    plt.title(str(unique_house_type[i]))
plt.show()    
```
